### This script is for molecular seletion and molecular visualization after IE and EA prediction

```python
import pandas as pd
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw

data = pd.read_csv("result_zinc_30_RNN.csv")    

ls_smi = data[(data['IE'] > 8.44) & (data['EA'] < 0.54)]['smiles'].tolist()

ls_mol = [Chem.MolFromSmiles(smi) for smi in ls_smi]
img=Draw.MolsToGridImage(ls_mol,molsPerRow=4,subImgSize=(200,200),legends=[smi for smi in ls_smi], useSVG = True)
```
